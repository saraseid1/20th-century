```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from pyvis.network import Network
from cdlib import algorithms
```

```python
import pandas as pd

df = pd.read_csv("country_relationships.csv")
df.head()
```

```python
import ast

# Convert string list into actual Python list
df["Countries_in_Sentence"] = df["Countries_in_Sentence"].apply(ast.literal_eval)

relationships = []

for row in df["Countries_in_Sentence"]:
    # create pair combinations
    for i in range(len(row)):
        for j in range(i+1, len(row)):
            relationships.append({
                "source": row[i],
                "target": row[j],
                "value": 1
            })

# Convert to dataframe
edges_df = pd.DataFrame(relationships)

# Combine duplicate pairs by summing value
edges_df = edges_df.groupby(["source","target"], as_index=False).sum()

edges_df.head()
```

```python
import networkx as nx

G = nx.from_pandas_edgelist(
    edges_df,
    source="source",
    target="target",
    edge_attr="value",
    create_using=nx.Graph()
)

G
```

```python
import matplotlib.pyplot as plt
import networkx as nx

plt.figure(figsize=(12, 12))

# Use a layout tp position nodes
pos = nx.kamada_kawai_layout(G)

# Draw the graph
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=800,
    node_color="skyblue",
    font_size=10,
    edge_color="gray"
)

plt.title("Static Network Graph – Country Relationships (20th Century)")
plt.show()
```

```python
from pyvis.network import Network

net = Network(notebook=True)
net.from_nx(G)
net.show("countries_network.html")
```

```python
from cdlib import algorithms

# Apply Louvain community detection
communities = algorithms.louvain(G)

communities
```

```python
new_com = communities.to_node_community_map()

dict_com = {k: v[0] for k, v in new_com.items()}
```

```python
nx.set_node_attributes(G, dict_com, "group")
```

```python
from pyvis.network import Network

net = Network(notebook=True, width="1100px", height="800px",
              bgcolor="#222222", font_color="white")

net.from_nx(G)
net.show("country_communities.html")
```

```python
from pyvis.network import Network

com_net = Network(notebook=True, width="1100px", height="800px", bgcolor="#222222", font_color="white")

com_net.from_nx(G)
com_net.show("country_communities.html")
```

The Leiden algorithm divided the countries into multiple communities based on how often they appeared together in the text. This partition makes sense historically because countries that were involved in the same events or conflicts tend to be grouped together. Nations that frequently interacted during major 20th-century events (alliances, wars, treaties, or political blocs) appear in the same community.
Overall, the communities show patterns of cooperation, opposition, and shared history. Countries with fewer mentions or weaker links appear in smaller or separate groups. These clusters help highlight which nations were most closely connected during major events of the century.

```python
# DEGREE CENTRALITY
degree_dict = nx.degree_centrality(G)
deg_df = pd.DataFrame.from_dict(degree_dict, orient='index', columns=['degree'])
deg_df = deg_df.sort_values('degree', ascending=False)

plt.figure(figsize=(12,6))
deg_df['degree'].plot(kind='bar', color='skyblue')
plt.title("Degree Centrality")
plt.ylabel("Centrality Score")
plt.xticks(rotation=90)
plt.show()
```

```python
# CLOSENESS CENTRALITY
closeness_dict = nx.closeness_centrality(G)
close_df = pd.DataFrame.from_dict(closeness_dict, orient='index', columns=['closeness'])
close_df = close_df.sort_values('closeness', ascending=False)

plt.figure(figsize=(12,6))
close_df['closeness'].plot(kind='bar', color='orange')
plt.title("Closeness Centrality")
plt.ylabel("Centrality Score")
plt.xticks(rotation=90)
plt.show()
```

```python
# BETWEenness CENTRALITY
betweenness_dict = nx.betweenness_centrality(G)
betw_df = pd.DataFrame.from_dict(betweenness_dict, orient='index', columns=['betweenness'])
betw_df = betw_df.sort_values('betweenness', ascending=False)

plt.figure(figsize=(12,6))
betw_df['betweenness'].plot(kind='bar', color='green')
plt.title("Betweenness Centrality")
plt.ylabel("Centrality Score")
plt.xticks(rotation=90)
plt.show()
```

 Degree Centrality
A few countries have many direct connections, meaning they appear in many historical relationships.
The highest-degree nodes likely represent major global powers or regions frequently involved in conflicts or alliances.
Most countries have low degree scores, showing they appear less often in the dataset.
 Closeness Centrality
Countries with the highest closeness scores are centrally located in the network, meaning they can reach many others through short paths.
These nodes act as important hubs in the historical context.
Peripheral countries have low closeness, meaning they are more isolated in the network.
 Betweenness Centrality
Only a few countries have high betweenness, meaning they act as bridges connecting different communities.
These nodes often link separate regions or political groups.
Most nodes have nearly zero betweenness, meaning they do not play a “middleman” role in the network.
