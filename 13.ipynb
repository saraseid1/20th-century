```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan'],
    'Age': np.random.randint(20, 50, size=5)
}
df = pd.DataFrame(data)
df
```

```python
 #Create  column 'Age_plus_5' by adding 5 to each Age value using a lambda function
df['Age_plus_5'] = df['Age'].apply(lambda x: x + 5)

df
```
