### Step 2: Key Elements I Will Focus On

After reviewing the Wikipedia page “Timeline of the 20th Century”,  
I found that it contains a list of major events with years and short descriptions.

For my project, I will focus on:
- The **year or decade** of each event  
- The **event title or description**  
- The **countries** mentioned in each event (to find relationships)  
- The **type of connection** between countries 

```python
# Import libraries
import requests
from bs4 import BeautifulSoup

# Define URL of the page you want to scrape
page_url = "https://en.wikipedia.org/wiki/Timeline_of_the_20th_century"

# Get page content
page = requests.get(page_url)

# Parse the HTML content
soup = BeautifulSoup(page.text, 'html.parser')

# Extract text from  page
text = soup.get_text()

# Encode and save the text to a .txt file
with open('20th_century_events.txt', 'wb') as f:
    f.write(text.encode('utf-8'))

print("✅ Page scraped successfully and saved as '20th_century_events.txt'")
```
